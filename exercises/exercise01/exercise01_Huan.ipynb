```python
# Q1：How many times did each type of request get made?
#Identify different types of request and extract their names
diff_request_types = []
for request in service_requests:
    request_type = request['Request Type']
    if request_type not in diff_request_types:
        diff_request_types.append(request_type)
print(diff_request_types)    
```

```python
#Iterate through each request types and count numbers respectively
type_count = {}
for type in diff_request_types:
    count = 0
    for request in service_requests:
        if request['Request Type'] == type:
            count += 1
    type_count[type] = count

print(type_count)
```

```python
# Find what types are most common
# Sort count
sorted_request_types = sorted(type_count.items(), key=lambda x: x[1], reverse=True)
print(sorted_request_types)
```

```python
#Q2: How many times did requests get made from each city?
#Identify different cities and extract their names
diff_cities = []
for request in service_requests:
    city = request['City']
    if city not in diff_cities:
        diff_cities.append(city)
print(diff_cities)
```

```python
#Iterate through each city and count numbers
city_count = {}
for city in diff_cities:
    count1 = 0
    for request in service_requests:
        if request['City'] == city:
            count1 += 1
    city_count[city] = count1

print(city_count)
```

```python
# Find what cities request often?
# Sort count
sorted_request_cities = sorted(city_count.items(), key=lambda x: x[1], reverse=True)
print(sorted_request_cities)
```
